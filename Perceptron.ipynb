```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X, y = make_classification(n_samples=1000, n_features=2, n_informative=1, n_redundant=0,n_classes=2,n_clusters_per_class=1, random_state=41)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette='Set1')
plt.title('Dataset Visualization')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
class Perceptron():
    def __init__(self,learning_rate=0.01,epochs=1000):
        self.learning_rate=learning_rate
        self.epochs=epochs
        self.weights=np.zeros(X.shape[1]+1)
        self.bias=1

    def fit(self,X,y):
        X=np.insert(X,0,1,axis=1)
        for i in range(self.epochs):
            random_point=np.random.randint(X.shape[0])
            y_hat=np.dot(X[random_point],self.weights)
            self.weights=self.weights+(self.learning_rate*(y[random_point]-step(y_hat))*X[random_point])

    def predict(self,X):
        X=np.insert(X,0,1,axis=1)
        preds=[]
        for i in range(X.shape[0]):
            preds.append(step(np.dot(X[i],self.weights)))
        return preds
def step(Xpreds):
    if Xpreds>0:
        return 1
    else:
        return 0
```

```python
X_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=41)
pr=Perceptron()
pr.fit(X_train,y_train)
```

```python
intercept=pr.weights[0]
slope=pr.weights[1:]
m=-(slope[0]/slope[1])
b=intercept/slope[1]
print(m,b)
```

```python
x_input=np.linspace(min(X[:, 0]), max(X[:, 0]), 100)
y_input=m*x_input+b
```

```python
ypreds=pr.predict(x_test)
```

```python
accuracy_score(y_test,ypreds)
```
